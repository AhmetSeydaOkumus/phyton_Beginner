```python
import time

my_Isim=["ali","hasan","mehmet","cemal"]
a=len(my_Isim)-1

while True:
    b=my_Isim[a].title()
    print(b)
    time.sleep(1)        #time kutuphanesinden kod sonucu yavaslatiyor
    a-=1
    if a==-1:
        break
```
